```python
def parse_ingredients(ingredients, quantities):
    ingredients_list = ingredients.split(',')
    quantities_list = [float(q) for q in quantities.strip().split(',')]  # Use float instead of int
    # Return a dictionary where keys are ingredient names and values are quantities
    return  dict(zip(ingredients_list, quantities_list))
```

```python

# Aggregate function to combine ingredients
def aggregate_ingredients(ingredient_lists):
    aggregated = {}
    # Iterate through each ingredient list
    for ing_list in ingredient_lists:
        for ing, qty in ing_list.items():
            if ing in aggregated:
                # If the ingredient is already in the dictionary, add the quantity
                aggregated[ing] += qty
            else:
                # Otherwise, set the initial quantity
                aggregated[ing] = qty
    return aggregated
```
