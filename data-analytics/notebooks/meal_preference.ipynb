```python
import pandas as pd

# Load the dataset
df = pd.read_csv('../meal_with_preference.csv')
df.head()
```

```python
# in Meal time column some rows '/' seperated difference values so we need to split them and create new rows so that each split and add into list and add dataset again
df = df.assign(Meal_time=df['Meal Time'].str.split('/')).explode('Meal_time')
```

```python
# using Meal time colummn seperate the data into breakfast, lunch and dinner
breakfast_df = df[df['Meal_time'] == 'Breakfast']
lunch_df = df[df['Meal_time'] == 'Lunch']
dinner_df = df[df['Meal_time'] == 'Dinner']
```

```python
lunch_df.head()
```

```python
# Function to generate a single random meal combination
def generate_random_combination():
    breakfast = breakfast_df.sample(n=1).iloc[0]
    lunch = lunch_df.sample(n=1).iloc[0]
    dinner = dinner_df.sample(n=1).iloc[0]

    # Combine ingredients (unique values only)
    combined_ingredients = set(breakfast['Ingredients'].split(', ') +
                               lunch['Ingredients'].split(', ') +
                               dinner['Ingredients'].split(', '))

    # Calculate the total nutritional values
    total_nutrition = {
        'Total Energy (Kcal)': breakfast['Energy(Kcal)'] + lunch['Energy(Kcal)'] + dinner['Energy(Kcal)'],
        'Total Protein (g)': breakfast['Protein(g)'] + lunch['Protein(g)'] + dinner['Protein(g)'],
        'Total Fat (g)': breakfast['Total fat(g)'] + lunch['Total fat(g)'] + dinner['Total fat(g)'],
        'Total Carbohydrates (g)': breakfast['Carbohydrates(g)'] + lunch['Carbohydrates(g)'] + dinner['Carbohydrates(g)'],
        'Total Magnesium (mg)': breakfast['Magnesium(mg)'] + lunch['Magnesium(mg)'] + dinner['Magnesium(mg)'],
        'Total Sodium (mg)': breakfast['Sodium(mg)'] + lunch['Sodium(mg)'] + dinner['Sodium(mg)'],
        'Total Potassium (mg)': breakfast['Potassium(mg)'] + lunch['Potassium(mg)'] + dinner['Potassium(mg)'],
        'Total Saturated Fatty Acids (mg)': breakfast['Saturated Fatty Acids(mg)'] + lunch['Saturated Fatty Acids(mg)'] + dinner['Saturated Fatty Acids(mg)'],
        'Total Monounsaturated Fatty Acids (mg)': breakfast['Monounsaturated Fatty Acids(mg)'] + lunch['Monounsaturated Fatty Acids(mg)'] + dinner['Monounsaturated Fatty Acids(mg)'],
        'Total Polyunsaturated Fatty Acids (mg)': breakfast['Polyunsaturated Fatty Acids(mg)'] + lunch['Polyunsaturated Fatty Acids(mg)'] + dinner['Polyunsaturated Fatty Acids(mg)'],
        'Total Free Sugar (g)': breakfast['Free sugar(g)'] + lunch['Free sugar(g)'] + dinner['Free sugar(g)'],
        'Total Starch (g)': breakfast['Starch(g)'] + lunch['Starch(g)'] + dinner['Starch(g)']
    }

    return {
        'Breakfast': breakfast['Complete Meal'],
        'Lunch': lunch['Complete Meal'],
        'Dinner': dinner['Complete Meal'],
        'Combined Ingredients': ', '.join(combined_ingredients),
        **total_nutrition
    }

# Generate 1000 combinations
combinations = [generate_random_combination() for _ in range(1000)]

# Convert to DataFrame
combinations_df = pd.DataFrame(combinations)

combinations_df.head()
```
