```python
# load a dataset
df_meal = pd.read_csv('../Dataset - Meals.csv')
df_meal.head()
```

```python
# check null values
df_meal.isnull().sum()
```

```python
# remove all meal ingredients with null values in column Meal_Ingredient_Ids  
df_meal = df_meal.dropna(subset=['Meal_Ingredient_Ids'])
```

```python
# Global variable to count errors
num = 0

# Function to check validity of rows
def check_validity(row):
    global num  # Declare num as global to modify it
    print(row['Name'])
    ingredient_ids = row['Meal_Ingredient_Ids'].split(', ')
    ingredients = row['Meal_Ingredients_Names'].split(', ')
    if len(ingredient_ids) != len(ingredients):
       
        num = num + 1

# Apply the function to the dataframe
df_meal.apply(check_validity, axis=1)

# Print the final count of errors
print(num)
```

```python
# In this case i Want to calculate total nutrition values of each meal in df_combined_meal dataset as Follows 
# 1. Get the nutrition values of each ingredient from df_nut dataset math with Meal_Ingredient_Ids in df_combined_meal dataset
# 2. devide each nutrional value by 100 to get 1g value and multiply by each ingredient quantity in df_combined_meal dataset
# 3. sum all the values of each corresponding ingredient in each meal to get total nutrition values of each meal
# Give full code
# Give full code for this


# List of columns to calculate
nutritional_columns = [
    'Energy(Kcal)', 'Protein(g)', 'Total fat(g)', 'Carbohydrates(g)', 
    'Total Dietary Fibre(g)', 'Vitamin A(µg)', 'Vitamin D(µg)', 'Viatmin K(µg)', 
    'Vitamin E(mg)', 'Calcium(mg)', 'Phosphorus(mg)', 'Magnesium(mg)', 
    'Sodium(mg)', 'Potassium(mg)', 'Saturated Fatty Acids(mg)', 
    'Monounsaturated Fatty Acids(mg)', 'Polyunsaturated Fatty Acids(mg)', 
    'Free sugar(g)', 'Starch(g)'
]

# Function to calculate nutritional values
def calculate_total_nutrition(row):
    print("----------------------------------------------------------- ",row['Complete Meal'])
    
    ingredient_ids = row['Meal_Ingredient_Ids'].split(', ')
    quantities = list(map(float, row['Quantities'].split(', ')))

    # Initialize a dictionary to store total nutritional values
    total_nutrition = {col: 0 for col in nutritional_columns}
    # print(total_nutrition)

    i = 0

    # Iterate over each ingredient and calculate the nutritional values
    for ingredient_id in ingredient_ids:
        print(ingredient_id)
        if ingredient_id == '':
            continue
        # Get the nutritional values for the current ingredient
        matching_rows = df_nut[df_nut['Food_Code'] == ingredient_id]
    
        # Check if matching rows were found
        if matching_rows.empty:
        # Skip this iteration if no match is found
            print(f"No nutritional values found for ingredient ID: {ingredient_id}. Skipping...")
            continue  

        # Get the nutritional values for the current ingredient
        nutrition_values = df_nut[df_nut['Food_Code'] == ingredient_id].iloc[0]
        
        # in nutrition_values, the first 2 column are Food Code and Price Code, so we skip it
        nutrition_values = nutrition_values[3:]
        

        # Calculate the nutritional values based on the quantity
        for nutrient in total_nutrition.keys():
            print(nutrient)
            print(nutrition_values[nutrient])
            print(nutrition_values[nutrient].dtype)
            value_nutrient = float(nutrition_values[nutrient])
            print((value_nutrient / 100) * quantities[i])
            total_nutrition[nutrient] += (value_nutrient / 100) * quantities[i]
    i = i+1
    print("Total nutrition : ",total_nutrition)
    return pd.Series(total_nutrition)

# Apply the function to each row in df_combined_meal
nutritional_values_df = df_combined_meal.apply(calculate_total_nutrition, axis=1)

# Concatenate the original df_combined_meal with the new nutritional values
df_combined_meal_with_nutrition = pd.concat([df_combined_meal, nutritional_values_df], axis=1)






```
