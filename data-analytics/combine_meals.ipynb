```python
# Ensure 'CAT' column in side_meals is a string and handle NaN values
side_meals['CAT'] = side_meals['CAT'].astype(str).fillna('')

# Convert 'CAT' to list of integers for comparison
def convert_to_int_list(cat_str):
    if not cat_str:
        return []  # Handle empty strings (or NaNs converted to empty)
    return [int(x.strip()) for x in cat_str.split(',') if x.strip().isdigit()]
```

```python

side_meals['CAT_list'] = side_meals['CAT'].apply(convert_to_int_list)
```

```python
# in main meal CAT column remove all , and convert to int
main_meals['CAT'] = main_meals['CAT'].str.replace(',', '').astype(int)
```

```python
# Iterate over each main meal
for _, main_meal in main_meals.iterrows():
    main_id =int(main_meal['CAT'])  # Convert main_id to integer
    main_name = main_meal['Name']
    
    # Find side meals that contain this main meal's ID in their CAT_list
    related_side_meals = side_meals[side_meals['CAT_list'].apply(lambda x: main_id in x)]
    
    # For each related side meal, create a combination
    for _, side_meal in related_side_meals.iterrows():
        side_name = side_meal['Name']
        meal_combinations.append({
            'Main Meal': main_name,
            'Side Meal': side_name
        })
```

```python
import itertools
import pandas as pd
import random

# Assuming meal_combinations_df has already been created as in the previous code

# List of main meals to process
main_meals = [ 'Chicken Biryani', 'Vegetable Biryani', 'Fish Biryani']

# Initialize a list to store all new meal combinations
all_meal_combinations = []

# Loop through each main meal and generate random combinations
for main_meal in main_meals:
    # Filter for the specific main meal
    meal_combinations = meal_combinations_df[meal_combinations_df['Main Meal'] == main_meal]

    # Group by 'Main Meal' and collect relevant side meals
    side_meal_groups = meal_combinations.groupby('Main Meal')['Side Meal'].apply(list).reset_index()

    # Extract side meals related to the current main meal
    side_meals = side_meal_groups.loc[side_meal_groups['Main Meal'] == main_meal, 'Side Meal'].values[0]

    # Function to get random combinations
    def get_random_combinations(side_meals, r, n_samples=10):
        all_combinations = list(itertools.combinations(side_meals, r))
        sampled_combinations = random.sample(all_combinations, min(n_samples, len(all_combinations)))
        return sampled_combinations

    # Generate 100 random combinations of 3, 4, and 5 side meals for the current main meal
    for r in [3, 4]:
        sampled_combos = get_random_combinations(side_meals, r, n_samples=10)
        for combo in sampled_combos:
            all_meal_combinations.append({
                'Main Meal': main_meal,
                'Side Meal': list(combo)
            })

# Convert all meal combinations into a DataFrame
all_meal_combinations_df = pd.DataFrame(all_meal_combinations)

# Display the first 50 rows of the DataFrame
all_meal_combinations_df.head(50)

# Save the new meal combinations to a new CSV file (Optional)
# all_meal_combinations_df.to_csv('meal_combinations.csv', index=False)
```

```python
new2_meal_combinations_df['Ingredients'] = ''
new2_meal_combinations_df['Quantities'] = ''

for i in range(len(new2_meal_combinations_df)):
    # Get main meal and side meal names
    main_meal = new2_meal_combinations_df.loc[i, 'Main Meal']
    side_meal = new2_meal_combinations_df.loc[i, 'Side Meal']

    # Extract ingredients and quantities for the main meal
    main_meal_ingredients = df[df['Name'] == main_meal]['Meal_Ingredients_Names'].values[0].split(',')
    main_meal_quantities = df[df['Name'] == main_meal]['Meal_Ingredient_Quantities'].values[0].split(',')

    # Extract ingredients and quantities for the side meal
    side_meal_ingredients = df[df['Name'] == side_meal]['Meal_Ingredients_Names'].values[0].split(',')
    side_meal_quantities = df[df['Name'] == side_meal]['Meal_Ingredient_Quantities'].values[0].split(',')

    # Combine ingredients and quantities
    combined_ingredients = main_meal_ingredients + side_meal_ingredients
    combined_quantities = main_meal_quantities + side_meal_quantities

    # Initialize a dictionary to hold combined ingredients and their total quantities
    ingredient_dict = {}
    for ingredient, quantity in zip(combined_ingredients, combined_quantities):
        if ingredient in ingredient_dict:
            ingredient_dict[ingredient] += float(quantity)
        else:
            ingredient_dict[ingredient] = float(quantity)

    # Assign the combined ingredients and quantities to the new columns
    new2_meal_combinations_df.loc[i, 'Ingredients'] = ', '.join(ingredient_dict.keys())
    new2_meal_combinations_df.loc[i, 'Quantities'] = ', '.join([str(val) for val in ingredient_dict.values()])
    new2_meal_combinations_df.loc[i, 'Meal'] = f'{main_meal} + {side_meal}'

# Display the first 200 rows of the dataframe
new2_meal_combinations_df.head(200)
```
