### use standard scaler

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
prep_data=scaler.fit_transform(df.iloc[:,4:].to_numpy())
```

```python
prep_data[200:210]
```

### Training content based Filtering

```python
len(prep_data)
```

```python
## Training content based Filtering
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(metric='cosine',algorithm='brute')
neigh.fit(prep_data)
```

#### Make pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
transformer = FunctionTransformer(neigh.kneighbors,kw_args={'return_distance':False})
pipeline=Pipeline([('std_scaler',scaler),('NN',transformer)])
```

```python
params={'n_neighbors':5,'return_distance':False}
pipeline.get_params()
pipeline.set_params(NN__kw_args=params)
```

#### Try Predictions

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Define the fuzzy variables and their ranges
age = ctrl.Antecedent(np.arange(0, 101, 1), 'age')
weight = ctrl.Antecedent(np.arange(30, 201, 1), 'weight')
height = ctrl.Antecedent(np.arange(100, 251, 1), 'height')

# Define fuzzy output variables for nutrients
energy = ctrl.Consequent(np.arange(500, 3501, 1), 'energy')
protein = ctrl.Consequent(np.arange(0, 201, 1), 'protein')
fat = ctrl.Consequent(np.arange(0, 101, 1), 'fat')
carbohydrates = ctrl.Consequent(np.arange(0, 401, 1), 'carbohydrates')
magnesium = ctrl.Consequent(np.arange(0, 1001, 1), 'magnesium')
sodium = ctrl.Consequent(np.arange(0, 3001, 1), 'sodium')
potassium = ctrl.Consequent(np.arange(1000, 5001, 1), 'potassium')
saturated_fats = ctrl.Consequent(np.arange(0, 20001, 1), 'saturated_fats')
monounsaturated_fats = ctrl.Consequent(np.arange(0, 60001, 1), 'monounsaturated_fats')
polyunsaturated_fats = ctrl.Consequent(np.arange(0, 20001, 1), 'polyunsaturated_fats')
free_sugar = ctrl.Consequent(np.arange(0, 101, 1), 'free_sugar')
starch = ctrl.Consequent(np.arange(0, 151, 1), 'starch')

# Define membership functions for age, weight, height
age['young'] = fuzz.trapmf(age.universe, [0, 0, 20, 30])
age['middle_aged'] = fuzz.trapmf(age.universe, [25, 35, 50, 60])
age['senior'] = fuzz.trapmf(age.universe, [55, 65, 100, 100])

weight['underweight'] = fuzz.trapmf(weight.universe, [30, 30, 50, 60])
weight['normal'] = fuzz.trapmf(weight.universe, [55, 65, 75, 85])
weight['overweight'] = fuzz.trapmf(weight.universe, [80, 90, 110, 120])
weight['obese'] = fuzz.trapmf(weight.universe, [115, 130, 200, 200])

height['short'] = fuzz.trapmf(height.universe, [100, 100, 140, 155])
height['average'] = fuzz.trapmf(height.universe, [150, 165, 175, 185])
height['tall'] = fuzz.trapmf(height.universe, [180, 190, 250, 250])

# Define membership functions for nutritional outputs
energy['low'] = fuzz.trapmf(energy.universe, [500, 500, 1000, 1500])
energy['medium'] = fuzz.trapmf(energy.universe, [1400, 1800, 2500, 3000])
energy['high'] = fuzz.trapmf(energy.universe, [2800, 3200, 3500, 3500])

protein['low'] = fuzz.trapmf(protein.universe, [0, 0, 50, 75])
protein['medium'] = fuzz.trapmf(protein.universe, [60, 80, 120, 150])
protein['high'] = fuzz.trapmf(protein.universe, [140, 170, 200, 200])

# Repeat for other nutrients (fat, carbohydrates, etc.)
# Example for fat:
fat['low'] = fuzz.trapmf(fat.universe, [0, 0, 20, 30])
fat['medium'] = fuzz.trapmf(fat.universe, [25, 35, 55, 70])
fat['high'] = fuzz.trapmf(fat.universe, [65, 80, 100, 100])

# Define the fuzzy rules
rule1 = ctrl.Rule(age['young'] & weight['normal'] & height['average'],
                  (energy['medium'], protein['medium'], fat['medium'], carbohydrates['medium'], magnesium['medium'], sodium['medium'], potassium['medium'], saturated_fats['medium'], monounsaturated_fats['medium'], polyunsaturated_fats['medium'], free_sugar['medium'], starch['medium']))

# Define more rules here...

# Create the control system and simulation
food_ctrl = ctrl.ControlSystem([rule1])
food_simulation = ctrl.ControlSystemSimulation(food_ctrl)

# Get inputs from the user
user_input = {
    "age": int(input("Enter your age (0-100): ")),
    "weight": int(input("Enter your weight in kg (30-200): ")),
    "height": int(input("Enter your height in cm (100-250): "))
}

# Input the values into the simulation
food_simulation.input['age'] = user_input['age']
food_simulation.input['weight'] = user_input['weight']
food_simulation.input['height'] = user_input['height']

try:
    # Perform the simulation
    food_simulation.compute()

    # Output the recommendations
    print(f"\nRecommended Total Energy intake: {food_simulation.output['energy']:.2f} Kcal")
    print(f"Recommended Total Protein intake: {food_simulation.output['protein']:.2f} g")
    print(f"Recommended Total Fat intake: {food_simulation.output['fat']:.2f} g")
    # Repeat for other nutrients...
except ValueError as e:
    print("Error during computation: ", e)
```
