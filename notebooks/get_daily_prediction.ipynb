### Training content based Filtering


```python
len(prep_data)
```

```python
## Training content based Filtering
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(metric='cosine',algorithm='brute')
neigh.fit(prep_data)
```

#### Make pipeline


```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
transformer = FunctionTransformer(neigh.kneighbors,kw_args={'return_distance':False})
pipeline=Pipeline([('std_scaler',scaler),('NN',transformer)])
```

```python
params={'n_neighbors':5,'return_distance':False}
pipeline.get_params()
pipeline.set_params(NN__kw_args=params)
```

### Get Total Calories using BMR

```python
def calculate_bmr(weight, height, age, gender):
    if gender.lower() == 'male':
        return 10 * weight + 6.25 * height - 5 * age + 5
    elif gender.lower() == 'female':
        return 10 * weight + 6.25 * height - 5 * age - 161
    else:
        raise ValueError("Gender must be 'male' or 'female'")

def calculate_daily_calories(bmr, activity_level):
    activity_factors = {
        'sedentary': 1.2,
        'lightly active': 1.375,
        'moderately active': 1.55,
        'very active': 1.725,
        'extra active': 1.9
    }
    return bmr * activity_factors.get(activity_level.lower(), 1.2)

def main():
    weight = float(input("Enter your weight in kg: "))
    height = float(input("Enter your height in cm: "))
    age = int(input("Enter your age in years: "))
    gender = input("Enter your gender (male/female): ")
    activity_level = input("Enter your activity level (sedentary, lightly active, moderately active, very active, extra active): ")

    try:
        bmr = calculate_bmr(weight, height, age, gender)
        daily_calories = calculate_daily_calories(bmr, activity_level)
        print(f"Your BMR is: {bmr:.2f} calories/day")
        print(f"Your daily calorie needs are: {daily_calories:.2f} calories/day")
    except ValueError as e:
        print(e)
main()
```
